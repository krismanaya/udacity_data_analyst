```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize']=(10,10)
```

```python
### load the dataframe 
simpsons = 'simpsons_episodes.csv'
df = pd.read_csv(simpsons)
df.head()
```

```python
### gather the golden years
df_g = df[df['season'] <= 9]
```

```python
sns.boxplot('season', 'us_viewers_in_millions', data=df_g)
```
